## Problem Statement & Goal
The objective of this task is to learn how to load, inspect, and visualize a dataset to understand underlying data trends and distributions. This fulfills the requirement to establish a clear problem statement and goal for the analysis.

## Overview of Steps

Using the Iris Dataset, this notebook will walk through the foundational steps of data exploration:
* **Data Loading & Inspection:** Loading the dataset using `pandas` and inspecting its shape, column names, and initial rows
* **Descriptive Statistics:** Utilizing summary statistics to understand the numerical properties of the data.
* **Data Visualization:** Creating scatter plots to find relationships between features, histograms to show value distributions, and box plots to identify any outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')
```

# Step 1: Data Inspection

```python

#tells us the exact size of our dataset
print("Shape of dataset:", df.shape)

#prints out a list of all the column headers.
print("Columns:", df.columns)

#displays the first 5 rows of the dataset in a clean table format
df.head()
```

# Step 2: Check Statistics

```python

# Check data types and look for any missing (null) values
print("Dataset Info:")
df.info()

print("\n-----------------------------------\n")

# Generate descriptive statistics (mean, min, max, quartiles)
print("Summary Statistics:")
df.describe()
```

#Step 3: Data Visualization

```python
# Scatter Plot(Relationships between features)
#This shows how two features relate to each other. We'll compare Sepal Length and Sepal Width, and color-code them by species

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Relationship between Sepal Length and Width')
plt.show()
```

```python
# Histograms (Value distributions)
# This shows how the values of a specific feature are spread out.

plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='petal_length', kde=True, bins=20)
plt.title('Distribution of Petal Length')
plt.show()
```

```python
#Box Plots (Identifying outliers)
#Box plots are perfect for spotting data points that are unusually high or low

plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plot of Iris Dataset Features (Checking for Outliers)')
plt.show()
```
